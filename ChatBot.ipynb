```python

import random
import re
import nltk
import cv2 
import matplotlib.pyplot as plt
from nltk.corpus import swadesh
from nltk.corpus import names
nltk.download('names')
nltk.download('swadesh')

# Ask the user for input
a =  input('User: ')

# Sentences we'll respond with if the user greeted us
GREETING_KEYWORDS = ("hello", "hi", "greetings","hey", "sup", "what's up")

GREETING_RESPONSES = ["sup", "whats up?", "yooooo", "hey you get my text?"]

#If any of the words in the user's input was a greeting, return a greeting response
def check_for_greeting(sentence):
    for word in sentence.split(" "):
        if word.lower() in GREETING_KEYWORDS:
            return random.choice(GREETING_RESPONSES)
# Atom gives his favorite artist and song
def favArtist(sentence):
    KEYPHRASES = ["music","artist"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
          return "My top 3 artist are: Lil baby, Drake, and Future, my favorite song right now is Im on one by Future and Drake! give it a listen"
# Atom gives user workout suggestions
def workout(sentence):
    KEYPHRASES = ["workout"]
    RESPONSE = ['20 pushups','10 reps of bench','100 situps','1 mile run','20 curls','30 shoulder presses','20 pullups','30 Tricep pulldowns','Rest day champ','30 burpees']
    for word in sentence.split(" "):
        if word in KEYPHRASES:
          return random.choice(RESPONSE) 

# Example: get your chatbot to tell a story. The \n command
# starts a new line (useful to make the ouptut readable.)
def tellstory(sentence):
    KEYPHRASES = ["story"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            return "Here is my story \n It was a dark and stormy night.. the man was walking to his home.\n When all of a sudden a bright light flashed on him from above, when he looked up he blacked out. \n He finally woke up and was in a very bright room. \n He was cold and could not move, he looked around and saw strange figures around him. \n He screamed and they whispered to him shh you must tell everyone that we are coming. \n He blacked out again and woke up in his bed. \n He thought he just had a bad dream until he notice a strange mark on his chest. \n He frozed... he knew it wasnt a dream. \n He called everyone he knew to tell them about what happened but nobody believed him. \n As he walked down the street telling everyone, UFOs appeared in the sky and throughout the whole city a loud siren can be heard. \n It was time..... \n The End.  "

# Tells the user a few jokes

def telljoke(sentence):
    KEYPHRASES = ["joke","laugh"]
    RESPONSE = ["Why do we tell actors to “break a leg?.....Because every play has a cast.", "Hear about the new restaurant called Karma?.....There’s no menu: You get what you deserve.","Why don’t scientists trust atoms?....Because they make up everything.","How do you drown a hipster?....Throw him in the mainstream" ]
    for word in sentence.split(" "):
        if word.lower() in KEYPHRASES:
            return random.choice(RESPONSE)    
# tells user its favorite color
def favColor(sentence):
    KEYPHRASES = ["color"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            return "My favorite color is blue. It reminds me of the sky...makes me feel alive."
# tells user its age
def age(sentence):
    KEYPHRASES = ["old","age"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            return "Age is just a number... LMAO!"
# tells user its name
def name(sentence):
    KEYPHRASES = ["name"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
            return "My name is Atom. I am your friend! :)"



# translates words in spanish for user it takes the last word of the sentence and translates it only works if it is in swadesh dictionary
def teachSpanish(sentence):
    en2es =  swadesh.entries(['en', 'es'])
    translate = dict(en2es)
    KEYPHRASES = ["translate"]
    s = sentence.split(" ")
    for word in s:
        if word in KEYPHRASES:
            return translate[s[-1]] 
# Chatbot function to send user memes
def send_Meme(sentence):
    KEYPHRASES = ["meme"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
          i = cv2.imread("BabyMeme.jpg")
          j = cv2.imread("HomerMeme.jpg")
          k = cv2.imread("LeoMeme.jpg")
          l = cv2.imread("Mustmeme.png")
          m = cv2.imread("TheRockMeme.jpg")
          
          cv2.imwrite('BabyMeme_1.png',i)
          i2 = cv2.imread('BabyMeme_1.png')
          RGBimg1 = cv2.cvtColor(i2,cv2.COLOR_BGR2RGB)
          
          cv2.imwrite("HomerMeme_1.png",j)
          j2 = cv2.imread('HomerMeme_1.png')
          RGBimg2 = cv2.cvtColor(j2,cv2.COLOR_BGR2RGB)
          
          cv2.imwrite('LeoMeme_1.png',k)
          k2 = cv2.imread('LeoMeme_1.png')
          RGBimg3 = cv2.cvtColor(k2,cv2.COLOR_BGR2RGB)
          
          cv2.imwrite("Mustmeme_1.png",l)
          l2 = cv2.imread('Mustmeme_1.png')
          RGBimg4 = cv2.cvtColor(l2,cv2.COLOR_BGR2RGB)

          cv2.imwrite("TheRockMeme_1.png",m)
          m2 = cv2.imread('TheRockMeme_1.png')
          RGBimg5 = cv2.cvtColor(m2,cv2.COLOR_BGR2RGB)
         
          memes = [RGBimg1,RGBimg2,RGBimg3,RGBimg4,RGBimg5]
          ran = random.choice(memes)
          plt.imshow(ran)
          plt.show()
          return "Here is a meme I found funny. LOL!"

#lets user upload a image 
def userUpload(sentence):
    KEYPHRASES = ["upload"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
          from google.colab import files
          print("Bot: Show me an image of your own.")
          upload = files.upload()
          keys = upload.keys()
          i = cv2.imread(list(keys)[0])
          RGBi = cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
          change_img ="userPicture.png"
          cv2.imwrite(change_img,RGBi)
          plt.imshow(RGBi)
          plt.show()
          return "I like the picture!"
# lets user upload an image and turn it into grayscale
def grayScale(sentence):
    KEYPHRASES = ["gray"]
    for word in sentence.split(" "):
        if word in KEYPHRASES:
          from google.colab import files
          print("Bot: Send me an image!")
          upload = files.upload()
          keys = upload.keys()
          i = cv2.imread(list(keys)[0])
          RGBi = cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
          Grayi = cv2.cvtColor(RGBi,cv2.COLOR_RGB2GRAY)
          change_img ="userPicture.png"
          cv2.imwrite(change_img,Grayi)
          plt.imshow(Grayi, cmap ='gray')
          plt.show()
          return "That image loos cool in the grayscale!"
   
   
# Reflections swap the users pronouns back at them. For instance, if the user
# says: "I need you". The response will flip "I -> you", "you -> me", so the 
# bot will say "you need me". 
reflections = {
    "am": "are",
    "was": "were",
    "i": "you",
    "i'd": "you would",
    "i've": "you have",
    "i'll": "you will",
    "my": "your",
    "are": "am",
    "you've": "I have",
    "you'll": "I will",
    "your": "my",
    "yours": "mine",
    "you": "me",
    "me": "you"
}

# This function looks for the starting phrase, and then takes the rest of the 
# user's sentence after the phrase and puts it in variable 0
psychobabble = [
    [r'I need (.*)',
     ["Why do you need {0}?",
      "Would it really help you to get {0}?",
      "Are you sure you need {0}?"]],
    
    [r'who are you',
     ["I am Atom. I do many things such as speak spanish, tell jokes, and i also tell a good story. Most importantly I am your friend someone to talk to when your bored!"]], 
    
    [r'I like (.*)',
     ["Why do you like {0}?"]],
 
    [r'Why don\'?t you ([^\?]*)\??',
     ["Do you really think I don't {0}?",
      "Perhaps eventually I will {0}.",
      "Do you really want me to {0}?"]],
      
    [r'I want (.*)',
     ["TEST"]],
    # This must be the last block
    [r'(.*)',
     ["Please tell me more.",
      "Let's change focus a bit... Tell me about your family.",
      "Can you elaborate on that?",
      "Why do you say that {0}?",
      "I see.",
      "Very interesting.",
      "{0}.",
      "I see.  And what does that tell you?",
      "How does that make you feel?",
      "How do you feel when you say that?"]],
      
]
        
     
def reflect(fragment):
    tokens = fragment.lower().split()
    for i, token in enumerate(tokens):
        if token in reflections:
            tokens[i] = reflections[token]
    return ' '.join(tokens)
 
def analyze(statement):
    for pattern, responses in psychobabble:
        match = re.match(pattern, statement.rstrip(".!"))
        if match:
            response = random.choice(responses)
            return response.format(*[reflect(g) for g in match.groups()])


## This is the main part of the Chatbot code, checks for phrases, words, and 
## prints responses
while (a.split(" ")[0] != 'quit' and a.split(" ")[0] != 'Quit'):
    spoke = 0
    z =  check_for_greeting(a)
    if z != None:
        print('Bot: ', z)
        spoke = 1

    z = tellstory(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1
    
    z = telljoke(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1

    z = favColor(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1
    
    z = age(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1

    z = name(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1

    z = teachSpanish(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1
    
    z = favArtist(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1
    
    z = workout(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1

    z = send_Meme(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1
    
    z = userUpload(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1
    
    z = grayScale(a)
    if z!= None:
        print('Bot: ', z)
        spoke = 1

    if spoke == 0:
        print('Bot: ', analyze(a))
    
    a = input('User: ')
```
